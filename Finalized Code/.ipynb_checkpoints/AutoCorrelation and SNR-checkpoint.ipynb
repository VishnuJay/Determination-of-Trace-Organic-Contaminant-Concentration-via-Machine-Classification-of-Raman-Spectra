```python
import pandas as pd
import numpy as np
from scipy import stats as stats
import scipy.signal as sgn
import scipy.integrate as aderiv
import matplotlib.pyplot as plt
import seaborn as sb
import KeyFunctions as me
```

```python
Rdf, Rlabels = me.ConstructCombinedDataset()
Tdf, Tlabels = me.ConstructCombinedTriclosanDataset()
Cdf, Clabels = me.ConstructCombinedChlorDataset()
Rdf = me.BaselineCorrection(me.BaselineCorrection(Rdf))
Tdf = me.BaselineCorrection(me.BaselineCorrection(Tdf))
Cdf = me.BaselineCorrection(me.BaselineCorrection(Cdf))
```

```python
RdfA = Rdf.reset_index()
Rauto = pd.DataFrame(columns = ['Concentration', 'AutoCorr'], dtype = 'float64')
for ind, row in RdfA.iterrows():
    conc = row[0]
    series = row[1:].astype(float)
    Rauto.loc[ind, 'AutoCorr'] = series.autocorr(1)
    Rauto.loc[ind, 'Concentration'] = conc
idx = Rauto['Concentration'].str.split('-', 1, expand = True).astype(int).drop(0, axis = 1).sort_values(1).index
Rauto = Rauto.reindex(index = idx)
Rauto = Rauto.reset_index(drop = True)

RautoAVG = Rauto.groupby('Concentration').mean()
RautoAVG = RautoAVG.reset_index()
idx = RautoAVG['Concentration'].str.split('-', 1, expand = True).astype(int).drop(0, axis = 1).sort_values(1).index
RautoAVG = RautoAVG.reindex(index = idx)
RautoAVG = RautoAVG.reset_index(drop = True)
RautoAVG = RautoAVG.set_index('Concentration', drop = True)
Rlabels
```

```python
fig, ax = plt.subplots(2, 4, sharex = True, sharey = True, figsize = [20, 10])
plt.subplots_adjust(wspace = 0, hspace = 0.2)
for i, conc in enumerate(Rlabels):
    for ind, row in RdfA.iterrows():
        if row[0] == conc:
            pd.plotting.autocorrelation_plot(row[1:], ax = ax[int(i/4), i%4])
            ax[int(i/4), i%4].set_title(conc)
            ax[int(i/4), i%4].grid(False)
            if i%4 != 0:
                ax[int(i/4), i%4].set_ylabel('')         
```

```python
fig, ax = plt.subplots(2, 3, sharex = True, sharey = True, figsize = [20, 10])
plt.subplots_adjust(wspace = 0, hspace = 0.2)
for i, conc in enumerate(Clabels):
    for ind, row in CdfA.iterrows():
        if row[0] == conc:
            pd.plotting.autocorrelation_plot(row[1:], ax = ax[int(i/3), i%3])
            ax[int(i/3), i%3].set_title(conc)
            ax[int(i/3), i%3].grid(False)
            if i%3 != 0:
                ax[int(i/3), i%3].set_ylabel('')    
```

```python
Rints = pd.DataFrame(columns = ['Conc', 'SN', 'N', 'S'], index = RdfA.index)
for inddf, row in RdfA.iterrows():
    Rints.loc[inddf, 'Conc'] = row[0]
    row = row[1:].astype(float)
    peaks, temp = sgn.find_peaks(row, prominence = 0.1*max(row))
    width, heights, left, right = sgn.peak_widths(row, peaks,  rel_height = 0.75)
    
    x = row.index.astype(float)
    Rints.loc[inddf, 'SN'] = -aderiv.trapezoid(row, x = x)

    ranges = np.stack([row.index[left.astype(int)], row.index[right.astype(int)]], axis = 1)
    nrow = row.copy()
    for indrow, val in row.iteritems():
        count = 0
        for i in ranges:
            if (i[1]<= indrow) and (indrow <=i[0]):
                nrow[indrow] = heights[count]
            count += 1

    Rints.loc[inddf, 'N'] = -aderiv.trapezoid(nrow, x = x)
    Rints.loc[inddf, 'S'] = Rints.loc[inddf, 'SN']-Rints.loc[inddf, 'N']

idx = Rints['Conc'].str.split('-', 1, expand = True).astype(int).drop(0, axis = 1).sort_values(1).index
Rints = Rints.reindex(index = idx)
Rints = Rints.reset_index(drop = True)
Rints['S/SN'] = Rints['S']/Rints['SN']
Rints['N/SN'] = Rints['N']/Rints['SN']
Rints
```
