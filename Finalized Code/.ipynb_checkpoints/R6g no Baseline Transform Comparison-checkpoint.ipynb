```python
import pandas as pd
import numpy as np
from scipy import stats as stats
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import Perceptron
from sklearn.svm import LinearSVC
from sklearn.neighbors import KNeighborsClassifier
import KeyFunctions as me
```

```python
#Get Evaporating Ouzo Data
concs = ['10-5', '10-9', '10-11', '10-14', '10-16']
df = me.GetFolderSpectra("ESMA", "r6g(Evap Ouzo)", concs)
col_names = np.array(df.columns)
```

```python
#Get and Process Ag Rings Data
from scipy import signal

Source = "ESMA"
Chemical = "r6g(Ag Nano Rings)"

dfA1 = pd.DataFrame()
for j in [6,9]:
    Concentration = "10-" + str(j)
    for i in [1, 2, 3, 4, 5]:
        temp = me.GetSpectra(Source, Chemical, Concentration, str(i))
        temp.set_index("#Wave", inplace = True)
        targ_len = len(temp)
        temp = temp.transpose()
        temp = temp.rename({'#Intensity':Concentration}, axis = 0)
        temp.set_axis(np.array(range(targ_len)), axis =1, inplace = True)
        dfA1 = pd.concat([dfA1, temp], axis = 0)  
        
for j in [5,7,8]:
    Concentration = "10-" + str(j)
    for i in [1, 2, 3, 4, 5]:
        temp = me.GetSpectra(Source, Chemical, Concentration, str(i))
        temp.set_index("#Wave", inplace = True)
        temp = np.flip(signal.resample(temp, targ_len))
        temp = np.reshape(temp, [targ_len])
        temp = pd.DataFrame(temp)
        temp = temp.transpose()
        temp = temp.rename({0:Concentration}, axis = 0)
        temp.set_axis(np.array(range(targ_len)), axis =1, inplace = True)
        dfA1 = pd.concat([dfA1, temp], axis = 0) 
        
dfA1.set_axis(col_names, axis =1, inplace = True)
dfA1
```

```python
#Transform Ouzo Data
dfO = pd.DataFrame()
for k in range(len(df)):
    Concentration = df.index[k]
    temp = pd.DataFrame(fwht(df.iloc[k,:])).transpose()
    temp = temp.rename({0:Concentration}, axis = 0)
    dfO = pd.concat([dfO, temp], axis = 0)
    
dfO
```

```python
#Combine Datasets and Perorm Train Test Split

df = pd.concat([dfO, dfA], axis = 0)
df = df.fillna(0)

from sklearn.model_selection import train_test_split

[train, test] = train_test_split(df, random_state = 42, shuffle = True, train_size = 0.8)

y_tn = train.index
y_tt = test.index
X_tt = test.to_numpy()
X_tn = train.to_numpy()
```

```python
from sklearn.preprocessing import StandardScaler

Scale = StandardScaler()

X_tnh = Scale.fit_transform(X_tn)
X_tth = Scale.transform(X_tt)
X_tnh
y_tn
```

```python
LinClass = Perceptron()
LinClass.fit(X_tnh, y_tn)
ypred = LinClass.predict(X_tth)

Lincounter = 0

for i in range(len(y_tt)):
    if ypred[i] != y_tt[i]:
        display("Incorrect Prediction at " + str(i) + ", Predicted: "+ str(ypred[i]) + "; Actual: " + str(y_tt[i]))
        Lincounter += 1
                
display("Total Predictions: " + str(i))
display("Total Errors: " + str(Lincounter))

RFC = RandomForestClassifier(random_state = 42)
RFC.fit(X_tnh, y_tn)
ypred = RFC.predict(X_tth)

RFcounter = 0

for i in range(len(y_tt)):
    if ypred[i] != y_tt[i]:
        display("Incorrect Prediction at " + str(i) + ", Predicted: "+ str(ypred[i]) + "; Actual: " + str(y_tt[i]))
        RFcounter += 1
                
display("Total Predictions: " + str(i))
display("Total Errors: " + str(RFcounter))

SVC = LinearSVC(max_iter = 10000, random_state = 42)
SVC.fit(X_tnh, y_tn)
ypred = SVC.predict(X_tth)

SVcounter = 0

for i in range(len(y_tt)):
    if ypred[i] != y_tt[i]:
        display("Incorrect Prediction at " + str(i) + ", Predicted: "+ str(ypred[i]) + "; Actual: " + str(y_tt[i]))
        SVcounter += 1
                
display("Total Predictions: " + str(i))
display("Total Errors: " + str(SVcounter))


KNN = KNeighborsClassifier()
KNN.fit(X_tnh, y_tn)
ypred = KNN.predict(X_tth)

KNcounter = 0

for i in range(len(y_tt)):
    if ypred[i] != y_tt[i]:
        display("Incorrect Prediction at " + str(i) + ", Predicted: "+ str(ypred[i]) + "; Actual: " + str(y_tt[i]))
        KNcounter += 1
                
display("Total Predictions: " + str(i))
display("Total Errors: " + str(KNcounter))
```

```python
from sklearn.model_selection import cross_val_score

df.columns = df.columns.astype(str)

X_h = np.append(X_tnh, X_tth, axis = 0)
y = np.append(y_tn, y_tt)

RFCscore = cross_val_score(RFC, X_h, y)
display("Random Forest Cross Validation")
display(RFCscore)
display(np.mean(RFCscore))

SVCscore = cross_val_score(SVC, X_h, y)
display("Support Vector Cross Validation")
display(SVCscore)
display(np.mean(SVCscore))

KNCscore = cross_val_score(KNN, X_h, y)
display("K-Neighbors Cross Validation")
display(KNCscore)
display(np.mean(KNCscore))
```

```python
from sklearn.metrics import confusion_matrix

labels = df.index.unique().values.tolist()
labels.sort()
labels.insert(-1, labels.pop(0))

RFcf = np.zeros([len(labels),len(labels)])
SVcf = np.zeros([len(labels),len(labels)])
KNcf = np.zeros([len(labels),len(labels)])

for n in [3, 25, 42, 17, 101]:
    [train, test] = train_test_split(df, random_state = n, shuffle = True, train_size = 0.8)

    y_tnr = train.index
    y_ttr = test.index
    X_ttr = test.to_numpy()
    X_tnr = train.to_numpy()

    Scale = StandardScaler()
    X_tnhr = Scale.fit_transform(X_tnr)
    X_tthr = Scale.transform(X_ttr)

    RFC = RandomForestClassifier(random_state = n)
    RFC.fit(X_tnhr, y_tnr)
    ypred = RFC.predict(X_tthr)
    RFcf += confusion_matrix(y_ttr, ypred, labels = labels)


    SVC = LinearSVC(max_iter = 10000, random_state = n)
    SVC.fit(X_tnhr, y_tnr)
    ypred = SVC.predict(X_tthr)
    SVcf += confusion_matrix(y_ttr, ypred, labels = labels)
    
    KNN = KNeighborsClassifier()
    KNN.fit(X_tnhr, y_tnr)
    ypred = KNN.predict(X_tthr)
    KNcf += confusion_matrix(y_ttr, ypred, labels = labels)
    
RFcf = RFcf/5
SVcf = SVcf/5
KNcf = KNcf/5

display(RFcf)
display(SVcf)
display(KNcf)
```

```python
fig, axs = plt.subplots(1, 3, sharey = True, figsize = [15, 5])

plt.rcParams['xtick.bottom'] = plt.rcParams['xtick.labelbottom'] = False
plt.rcParams['xtick.top'] = plt.rcParams['xtick.labeltop'] = True

pcm = axs[0].pcolor(RFcf, edgecolors = 'k', cmap = 'gist_heat_r')
plt.gca().invert_yaxis()
axs[0].set_xticks(ticks = np.linspace(0.5, len(labels)-0.5, num = len(labels)), labels = labels)
axs[0].set_yticks(ticks = np.linspace(0.5, len(labels)-0.5, num = len(labels)), labels = labels)
axs[0].set_ylabel("Actual Condition")
axs[0].set_xlabel("Predicted Condition")
axs[0].xaxis.set_label_position('top') 
axs[0].set_title('Random Forest Classifier');

axs[1].pcolor(SVcf, edgecolors = 'k', cmap = 'gist_heat_r');
plt.gca().invert_yaxis()
axs[1].set_xticks(ticks = np.linspace(0.5, len(labels)-0.5, num = len(labels)), labels = labels)
axs[1].set_title('Support Vector Classifier');
axs[1].set_xlabel("Predicted Condition")
axs[1].xaxis.set_label_position('top')

axs[2].pcolor(KNcf, edgecolors = 'k', cmap = 'gist_heat_r')
plt.gca().invert_yaxis()
axs[2].set_xticks(ticks = np.linspace(0.5, len(labels)-0.5, num = len(labels)), labels = labels);
axs[2].set_title('K-Nearest Neighbors Classifier');
axs[2].set_xlabel("Predicted Condition")
axs[2].xaxis.set_label_position('top')

fig.colorbar(pcm, ax = axs[:], location = 'bottom', label = 'Average Number of Assignments');
```
