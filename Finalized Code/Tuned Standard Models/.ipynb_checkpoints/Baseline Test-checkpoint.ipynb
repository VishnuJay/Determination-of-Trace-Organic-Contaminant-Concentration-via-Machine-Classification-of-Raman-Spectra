```python
import pandas as pd
import numpy as np
from scipy import stats as stats
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import Perceptron
from sklearn.svm import LinearSVC
from sklearn.neighbors import KNeighborsClassifier
import KeyFunctions as me
```

```python
from sklearn.model_selection import train_test_split

#Import Full R6g Dataset
df, labels = me.ConstructCombinedDataset()
```

```python
def ScaleSpectra(row):
    nrow = np.linspace(0, 1, len(row))
    for ind, x in enumerate(row):
        nrow[ind] = x/max(row)
    return nrow
```

```python
from sympy.discrete.transforms import fwht, ifwht
df_bl = df[df.iloc[:, 0] > 500]
df_nbl = df[df.iloc[:, 0] <= 500]


arr_blh = np.apply_along_axis(fwht, axis=1, arr=df_bl)
arr_nblh = np.apply_along_axis(fwht, axis=1, arr=df_nbl)
arr_blh = np.apply_along_axis(ScaleSpectra, axis=1, arr=arr_blh)
arr_nblh = np.apply_along_axis(ScaleSpectra, axis=1, arr=arr_nblh)

df_blh = pd.DataFrame(arr_blh, index = df_bl.index)
df_nblh = pd.DataFrame(arr_nblh, index = df_nbl.index)
plt.plot(df_blh.iloc[1, :])
```

```python
fig, ax = plt.subplots(2, 2, figsize = [20, 12], sharey = True, sharex = True)
plt.subplots_adjust(hspace = 0.1, wspace = 0)

blhavg = df_blh.loc[df_blh.index[:] == '10-5'].mean(axis = 0)
nblhavg = df_nblh.loc[df_nblh.index[:] == '10-5'].mean(axis = 0)
ax[0,0].set_title("$10^{-5}$ Spectra with Baseline Separated")
ax[0,0].plot(blhavg, figure = fig)
ax[0,0].plot(nblhavg, figure = fig)
ax[0,0].legend(["Average of Spectra With Baseline", "Average of Spectra Without Baseline"]);
ax[0,0].annotate("Baseline Specific Peak", xy=(65, 0.08), xytext=(75, 0.5),
            arrowprops=dict(width= 2, fc = 'k'))
ax[0,0].annotate("Baseline Exaggerated Peak", xy=(382, -0.1), xytext=(250, -0.3),
            arrowprops=dict(width= 2, fc = 'k'))
ax[0,0].annotate("Baseline Specific Peak", xy=(448, 0.08), xytext=(425, 0.5),
            arrowprops=dict(width= 2, fc = 'k'))


blhavg = df_blh.loc[df_blh.index[:] == '10-9'].mean(axis = 0)
nblhavg = df_nblh.loc[df_nblh.index[:] == '10-9'].mean(axis = 0)
ax[0,1].set_title("$10^{-9}$ Spectra with Baseline Separated")
ax[0,1].plot(blhavg, figure = fig)
ax[0,1].plot(nblhavg, figure = fig)
ax[0,1].legend(["Average of Spectra With Baseline", "Average of Spectra Without Baseline"]);
ax[0,1].annotate("Baseline Specific Peak", xy=(65, 0.08), xytext=(75, 0.5),
            arrowprops=dict(width= 2, fc = 'k'))
ax[0,1].annotate("Baseline Exaggerated Peak", xy=(382, -0.1), xytext=(250, -0.3),
            arrowprops=dict(width= 2, fc = 'k'))
ax[0,1].annotate("Baseline Specific Peak", xy=(448, 0.08), xytext=(425, 0.5),
            arrowprops=dict(width= 2, fc = 'k'))

blhavg = df_blh.loc[df_blh.index[:] == '10-11'].mean(axis = 0)
nblhavg = df_nblh.loc[df_nblh.index[:] == '10-11'].mean(axis = 0)
ax[1,0].set_title("$10^{-11}$ Spectra with Baseline Separated")
ax[1,0].plot(blhavg, figure = fig)
ax[1,0].plot(nblhavg, figure = fig)
ax[1,0].legend(["Average of Spectra With Baseline", "Average of Spectra Without Baseline"]);
ax[1,0].annotate("Baseline Specific Peak", xy=(65, 0.08), xytext=(75, 0.5),
            arrowprops=dict(width= 2, fc = 'k'))
ax[1,0].annotate("Baseline Exaggerated Peak", xy=(382, -0.1), xytext=(250, -0.3),
            arrowprops=dict(width= 2, fc = 'k'))
ax[1,0].annotate("Baseline Specific Peak", xy=(448, 0.08), xytext=(425, 0.5),
            arrowprops=dict(width= 2, fc = 'k'))

blhavg = df_blh.loc[df_blh.index[:] == '10-14'].mean(axis = 0)
nblhavg = df_nblh.loc[df_nblh.index[:] == '10-14'].mean(axis = 0)
ax[1,1].set_title("$10^{-14}$ No Spectra with Baseline")
ax[1,1].plot(blhavg, figure = fig)
ax[1,1].plot(nblhavg, figure = fig)
ax[1,1].legend(["Average of Spectra With Baseline", "Average of Spectra Without Baseline"]);
ax[1,1].annotate("Baseline Specific Peak", xy=(65, 0.08), xytext=(75, 0.5),
            arrowprops=dict(width= 2, fc = 'k'))
ax[1,1].annotate("Baseline Exaggerated Peak", xy=(382, -0.1), xytext=(250, -0.3),
            arrowprops=dict(width= 2, fc = 'k'))
ax[1,1].annotate("Baseline Specific Peak", xy=(448, 0.08), xytext=(425, 0.5),
            arrowprops=dict(width= 2, fc = 'k'))
```

```python
dfO, labels = me.ConstructCombinedDataset(3)
dfO = dfO.loc[dfO.index[:] == '10-16']

arr_O = np.apply_along_axis(fwht, axis=1, arr=dfO)
arr_O = np.apply_along_axis(ScaleSpectra, axis=1, arr=arr_O)

df_O = pd.DataFrame(arr_O, index = dfO.index)
plt.plot(df_O.iloc[1, :])
```

```python
from scipy.stats import zscore

avgspec = dfO.mean(axis = 0)

SSE = []
res = 0
for ind1, row in dfO.iterrows():
    res = 0
    for ind2, val in np.ndenumerate(row.values):
        res = res + (val-avgspec.iloc[ind2[0]])**2
    SSE.append(res)

dfOC = dfO.copy()    
dfOC['SSE'] = SSE
dfOC.sort_values('SSE', inplace = True, ascending = False)

avgSSE = np.mean(SSE)
dfDeviant = dfOC.iloc[0:4]
dfNormal = dfOC.iloc[4:]

dfDeviant.drop('SSE', inplace= True, axis = 1)
dfNormal.drop('SSE', inplace= True, axis = 1)

arr_dev = np.apply_along_axis(fwht, axis=1, arr=dfDeviant)
arr_norm = np.apply_along_axis(fwht, axis=1, arr=dfNormal)
arr_dev = np.apply_along_axis(ScaleSpectra, axis=1, arr=arr_dev)
arr_norm = np.apply_along_axis(ScaleSpectra, axis=1, arr=arr_norm)

df_dev = pd.DataFrame(arr_dev, index = dfDeviant.index)
df_norm = pd.DataFrame(arr_norm, index = dfNormal.index)

df_dev
```

```python
fig = plt.figure(figsize = [14, 10])

plt.plot(df_dev.mean(axis = 0), figure = fig)
plt.plot(df_norm.mean(axis = 0), figure = fig)
plt.legend(['Spectra Furthest from Average', 'Spectra Closest to Average'])
```
